#### Normal Distribution
mean = median = mode

```python
s = pd.Series([1, 2, 3, 3, 3, 4, 5]) # normal distribution
pd.DataFrame([[s.mean(), s.median(), s.mode()]], columns=['mean', 'median', 'mode'], index=['n'])
```

#### Kernel Density Estimation - KDE
Tells what is the probability distribution

```python
s.plot.kde()
```

#### Cummulative Distribution Function - CDF
Tells the accumulative probability

```python
s.hist( cumulative = True ) # 50% of data is value 3
```

#### Skewness
Tells "how much" asymmetric the data is



```python
pd.Series([1, 1, 2]).skew() # left asymmetric
```
